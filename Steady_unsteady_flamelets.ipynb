minimum and maximum value of the unscaled progress variable

```python
with open('Archive-Myra-Nelissen/Chem1D with DD/H3/ScalePV3.dat', 'r') as read_obj, \
    open('ScalePV3_1.csv', 'w', newline='\n') as write_obj:
        p=0
        for line in read_obj:
            if p==5:
                #print(line.lstrip().rstrip().split('\t')[0],line.lstrip().rstrip().split('\t')[1],line.lstrip().rstrip().split('\t')[2])
                x=[float(line.lstrip().rstrip().split('\t')[0]),float(line.lstrip().rstrip().split('\t')[1]),float(line.lstrip().rstrip().split('\t')[2])]
                writer(write_obj).writerow(x)
            if p>=6:
                x=[float(line.lstrip().rstrip().split('\t ')[0]),float(line.lstrip().rstrip().split('\t ')[1]),float(line.lstrip().rstrip().split('\t ')[2])]
                writer(write_obj).writerow(x)
            p=p+1
```

plotting minimum and maximum value of the unscaled progress variable

```python
x1=pd.read_csv('ScalePV3_1.csv',header=None).values
z=x1[:,0]
pv_min=x1[:,1]
pv_max=x1[:,2]
z=np.array(z)# converting list to an array
#print(z) #check z converted to a array (optional)
z=z.reshape(len(z),1) #converting to a column array
#print(z) #check(optional)
plt.plot(z, pv_min, 'r.', markersize=10, label='minimum value PV') # plotting observations
plt.plot(z, pv_max,'b-',label='maximum value PV')  #plotting predictions
plt.ylim([0, 12])
plt.title('Minimum and maximum value of the Progress variable') #defining titles
plt.xlabel('z[-]')             #defining x-axis equivalen ratios
plt.ylabel('Progress Variable($Y_{H_{2}O}$)')       #y-axis sensor measurements
plt.legend(loc='upper right')   #adding a legend
plt.show() 
```

creation file steady flamelets csv file

```python
with open('steady.csv', 'w', newline='\n') as write_obj:
    M=1000/18.01528 # raemaerkts factor in order to unsure that unscaled progress variable will be a strictly increasing function with respect to mixture fraction
    #print(M)
    # reading steady flamelets
    for files in os.listdir('Archive-Myra-Nelissen/Chem1D with DD/H3/Steady'):# this is done for counting the number of Matlab files
        name=files
        print(name) # checking whether steady flamelet files are being read correctly
        if name.split('.')[0]=='yiend':
            with open('Archive-Myra-Nelissen/Chem1D with DD/H3/Steady/'+name, 'r') as read_obj:
                p=0
                for line in read_obj:
                    if p>=22:
                    #print(line)
                        x=[M*float(line.lstrip().rstrip().split(' ')[4]),float(line.lstrip().rstrip().split(' ')[18])] #writing unscaled progress variable and mixture fraction in new file
                        writer(write_obj).writerow(x) # save unscaled Progress variable and mixture fraction in a new file called 'unsteady.csv'
                    p=p+1
```

Creating unsteady flamelets csv file

```python
with open('unsteady.csv', 'w', newline='\n') as write_obj:
    M=1000/18.01528 # raemaerkts factor in order to unsure that unscaled progress variable will be a strictly increasing function with respect to mixture fraction
    #print(M)
    # reading unsteady flamelets
    for files in os.listdir('Archive-Myra-Nelissen/Chem1D with DD/H3/Unsteady'):
        name=files
        print(name) # checking whether unsteady flamelets files are being read correctly
        if name.split('00')[0]=='yi':
            with open('Archive-Myra-Nelissen/Chem1D with DD/H3/Unsteady/'+name, 'r') as read_obj:
                p=0
                for line in read_obj:
                    if p>=22:
                    #print(line)
                        x=[M*float(line.lstrip().rstrip().split(' ')[4]),float(line.lstrip().rstrip().split(' ')[18])] #writing unscaled progress variable and mixture fraction in new file
                        writer(write_obj).writerow(x)  # save unscaled Progress variable and mixture fraction in a new file called 'unsteady.csv'
                    p=p+1
```

Plotting steady and unsteady flamelet: Unscaled progress variable with respect to mixture fraction

```python
x3=pd.read_csv('steady.csv',header=None).values   # reading steady flamelets using pandas
x2=pd.read_csv('unsteady.csv',header=None).values # reading unsteady flamelets using pandas
z_st=x3[:,1]        # reading mixture fraction from steady flamelet file
z_unst=x2[:,1]      # reading mixture fraction from unsteady flamelet file
pv_st=x3[:,0]       # read unscaled progress variable from steady flamelet file
pv_unst=x2[:,0]     # read unscaled progress variable from unsteady flamelet file

plt.figure(figsize=(8, 6)) 
plt.plot(z_st, pv_st, 'b.', markersize=5, label='steady') # plotting steady flamelets
plt.plot(z_unst, pv_unst,'y--',label='unsteady')  #plotting unsteady flamelets
plt.ylim([0, 12])                                 # limiting y-axis
plt.title('Steady and Unsteady flamelets',fontsize=15) #defining titles
plt.xlabel('z[-]',fontsize=15)             #defining x-axis mixture fraction
plt.ylabel('Progress Variable($Y_{H_{2}O}$)',fontsize=15)       #y-axis unscaled progress variable
plt.legend(loc='upper right',fontsize=15)   #adding a legend
plt.show() 
```
